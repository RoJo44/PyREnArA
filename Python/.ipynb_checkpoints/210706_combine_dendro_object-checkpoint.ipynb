```python
def hide_current_axis(*args, **kwds):
    plt.gca().set_visible(False)
```

```python
def convert_tif_jpg (self):
    import cv2
    import glob, os, errno

    import os

    root = '../../Linsel_Data/' 
    sitelist = [ item for item in os.listdir(root) if os.path.isdir(os.path.join(root, item)) ]

    for i in sitelist:

        mydir = (root + str({}) + '/' + str({}) + '_outlines/').format(i,i)
        #check if directory exist, if not create it

        for fil in glob.glob(mydir + "*.tif"):
            image = cv2.imread(fil) 

            gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY) # convert to greyscale


            fil_fin = fil.split("_")
            del fil_fin[-2]
            fil_fin = '_'.join(fil_fin)

            #print (fil_fin[:-4])
            fil_export = fil_fin[:-4]+'.jpg'      
            #cv2.imshow('reger',image)
            #cv2.waitKey(0)
            cv2.imwrite(os.path.join(fil_export),image) # write to location with same name
```

## Object orientated solution 

```python
class dendro:
    def __init__(self,data_link,category,legend_title,head_list,exp_path):
        self.data_import(data_link)
        
        
        self.head_list = head_list
        self.category = category
        self.x, self.category_list = self.data_separate(category)
        self.x_category = self.x.copy()
        self.x_category.loc [:,str(self.category)] = pd.Series(self.category_list, index=self.x_category.index)
        
        self.x2, self.category_list_first, self.category_list_second = self.data_separate_twoclasses('typology','site')
    
    def data_import(self,data_link):
        df_1 = pd.read_csv(data_link)


        df_temp_2 = df_1.iloc[:,0:35]

        # extract the site name  from the artifact ID 
        df_temp_2['site'] = df_1.ID.str.split(pat="_").str[0]#.tolist()[:,[0]]#.iloc[:,[1]]

        # Import of second site specific dataset
        second = '../../Linsel_Data/data_reftypchr_merged.csv'
        df_2 = pd.read_csv(second)

        # merging of both datasets 
        df_temp_2 = pd.merge(df_temp_2,df_2, left_index=True, right_index=True)

        # filtering of the datasets according to an archaeological selection
        df_temp_2 = df_temp_2[df_temp_2['class'] == 1]
        df_temp_2 = df_temp_2.sort_values(by=['ka_cal_BP'])

        self.df_temp = df_temp_2
    
    def data_separate(self,category): # ,category,head_list):
        data_metrics = self.df_temp[self.head_list]

        data_category = self.df_temp[self.category] 
        return data_metrics, data_category
    
    
    def data_separate_twoclasses (self,category_1, category_2): 
        data_metrics = self.df_temp[self.head_list]

        data_category_first = self.df_temp[category_1] 
        data_category_second = self.df_temp[category_2]
        return data_metrics, data_category_first, data_category_second
    
    def print_dendro(self,exp_add):
        category_pal = sns.cubehelix_palette(self.category_list.unique().size,rot=-.4)
        category_lut = dict(zip(map(str, self.category_list.unique()), category_pal))
        category_colors = self.category_list.map(category_lut)


        g = sns.clustermap(self.y, 
                           row_colors=category_colors,
                           standard_scale=1,
                           yticklabels=False)

        for label in self.category_list.unique():
            g.ax_col_dendrogram.bar(0, 0, color=category_lut[label],
                                    label=label, linewidth=0)
        g.ax_col_dendrogram.legend(title=legend_title, loc=(1.2, -4.6), ncol=1,fontsize=10)
        g.fig.subplots_adjust(right=0.7,top=0.9)
        g.ax_cbar.set_position((1.1, .2, .05, .2))
        #g.ax_col_dendrogram.set_position((0.1, .2, .03, .2))

        g.savefig(exp_path + exp_add + '.png', dpi=1200, bbox_inches="tight")

    def print_dendro_twoclasses (self,exp_add):
        category_pal = sns.cubehelix_palette(self.category_list_first.unique().size,rot=-.4)
        category_lut = dict(zip(map(str, self.category_list_first.unique()), category_pal))
        category_colors = self.category_list_first.map(category_lut)
        #THIS STUFF IS NEW
        # HEMI COLORS
        category_second_pal = sns.cubehelix_palette(self.category_list_second.unique().size,rot=-.4)
        category_second_lut = dict(zip(map(str, self.category_list_second.unique()), category_second_pal))
        category_second_colors = self.category_list_second.map(category_second_lut)

        # CREATE HEMI INDEX
        site_typology_colors = pd.DataFrame(category_colors).join(pd.DataFrame(category_second_colors))
        
        g = sns.clustermap(self.y, 
                           row_colors=site_typology_colors,
                           standard_scale=1,
                           yticklabels=False)

        for label in self.category_list_first.unique():

            g.ax_row_dendrogram.bar(0, 0, color=category_lut[label],
                                    label=label, linewidth=0)
            
        l1 = g.ax_row_dendrogram.legend(title='Typology', loc=(5.25, .0), ncol=1,fontsize=10)#, bbox_to_anchor=(0.7,0.9))

        for label in self.category_list_second.unique():

            g.ax_col_dendrogram.bar(0, 0, color=category_second_lut[label],
                                    label=label, linewidth=0)
        l2 = g.ax_col_dendrogram.legend(title='Sites', loc=(1.49, -.705), ncol=1,fontsize=10)#, bbox_to_anchor=(1.47, 1))
   
        
        g.ax_cbar.set_position((1.35, .2, .05, .2))
        
        #g.ax_col_dendrogram.legend(title=legend_title, loc=(1.2, -4.6), ncol=1,fontsize=10)
        #g.fig.subplots_adjust(right=0.7,top=0.9)
        #g.ax_cbar.set_position((1.1, .2, .05, .2))
        #g.ax_col_dendrogram.set_position((0.1, .2, .03, .2))

        g.savefig(exp_path + exp_add + '.png', dpi=1200, bbox_inches="tight")
        
    def dendro_normalized (self):

        #x = self.df_temp.values #returns a numpy array
        min_max_scaler = preprocessing.MinMaxScaler()
        x_scaled = min_max_scaler.fit_transform(self.x)
        data_metrics_scaled = pd.DataFrame(x_scaled)
        data_metrics_scaled.columns  = head_list #['backing_right','bifacial_right','dorsal_right','ventral_right','backing_left','bifacial_left','dorsal_left','ventral_left']

        self.y = data_metrics_scaled
    
    def dendro_pca (self,pca_no,head_list_pca):
        
        x = self.x.values

        pca = PCA(n_components=pca_no)
        principalComponents = pca.fit_transform(x)

        data_metrics_pca = pd.DataFrame(principalComponents)
        data_metrics_pca.columns  = head_list_pca #['width','length','l-w-Index', 'area', 'percent_area', 'contour_length', 'length_lower_part', 'length_upper_part', 'width_left_part','width_right_part','distance_MP_CM', 'tip_angle']
        
        self.y = data_metrics_pca
    
    
    def manova_prep(self):
        for item in self.head_list:
            try:
                manova_str = manova_str + ' + ' + item
                continue
            except NameError:
                manova_str = item

        manova_str = manova_str + ' ~ ' + data_category
        return manova_str
    
    def manova(self, exp_add):
            manova_str = self.manova_prep ()
            self.df_temp = self.y
            
            self.df_temp[data_category] = self.category_list
            
            manov = MANOVA.from_formula(manova_str, data=self.df_temp)

            result_maov = pd.DataFrame(manov.mv_test().summary().tables[1])
            result_maov = result_maov.append(manov.mv_test().summary().tables[3])
            result_maov.to_csv(exp_path + '_' + exp_add +'_manova.csv')
            
    def pairsplot_pearson (self,exp_add):

        #x_category = self.x
        #x_category.loc [:,str(self.category)] = pd.Series(self.category_list, index=x_category.index)
        e = sns.pairplot(data=self.x_category, hue=str(self.category))

        e.map_upper(hide_current_axis)

        (xmin, _), (_, ymax) = e.axes[0, 0].get_position().get_points()
        (_, ymin), (xmax, _) = e.axes[-1, -1].get_position().get_points()

        ax = e.fig.add_axes([xmin, ymin, xmax - xmin, ymax - ymin], facecolor='none')

        corr1 = self.x_category.corr()
        mask1 = np.tril(np.ones_like(corr1, dtype=bool))
        sns.heatmap(corr1, mask=mask1, cmap='seismic', vmax=.5, vmin=-.5,
                    linewidths=.5, cbar=False, annot=True, annot_kws={'size': 22}, ax=ax)
        ax.set_xticks([])
        ax.set_yticks([])
        # ax.xaxis.tick_top()
        # ax.yaxis.tick_right()


        e.savefig(exp_path + '_' + exp_add +'_pairsplotpearson.png')

        plt.show()

    def LDA_2(self):
        from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
        from sklearn.preprocessing import LabelEncoder
        from sklearn.tree import DecisionTreeClassifier
        from sklearn.model_selection import train_test_split
        from sklearn.metrics import confusion_matrix
        from sklearn.metrics import accuracy_score
        
        le = LabelEncoder()
        y = le.fit_transform(self.category_list.values)
        lda = LinearDiscriminantAnalysis()
        X_train, X_test, y_train, y_test = train_test_split(self.x, y, random_state=1)
        
        X_lda = lda.fit(self.x, y)
        
        y_pred = X_lda.predict(X_test)
        array = confusion_matrix(y_test, y_pred)
        print(array)
        
        df_cm = pd.DataFrame(array)#, index = self.category_list.unique(),
                  #columns = self.category_list.unique())
        plt.figure(figsize = (10,7))
        sns.heatmap(df_cm, annot=True)
        print('Accuracy' + str(accuracy_score(y_test, y_pred)))
        

        plt.xlabel('LD1 (' +')')
        plt.ylabel('LD2 (' + ')')
        for i in self.category_list.unique():
            plt.scatter(
                X_lda[:,0],
                X_lda[:,1],
                c=y,
                label = i,
                cmap='rainbow',
                alpha=0.7,
                edgecolors='b'
            )
        plt.legend(loc="upper left")
        plt.show()        
        
    def LDA(self):
        from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
        from sklearn.preprocessing import LabelEncoder
        from sklearn.tree import DecisionTreeClassifier
        from sklearn.model_selection import train_test_split
        from sklearn.metrics import confusion_matrix
        from sklearn.metrics import accuracy_score
        
        lda = LinearDiscriminantAnalysis()
        le = LabelEncoder()
        y = le.fit_transform(self.category_list.values)
        X_lda = lda.fit_transform(self.x, y)
        lda.explained_variance_ratio_

        plt.xlabel('LD1 (' +')')
        plt.ylabel('LD2 (' + ')')
        for i in self.category_list.unique():
            plt.scatter(
                X_lda[:,0],
                X_lda[:,1],
                c=y,
                label = i,
                cmap='rainbow',
                alpha=0.7,
                edgecolors='b'
            )
        plt.legend(loc="upper left")
        plt.show()
        X_train, X_test, y_train, y_test = train_test_split(X_lda, y, random_state=1)
        dt = DecisionTreeClassifier()
        dt.fit(X_train, y_train)
        y_pred = dt.predict(X_test)
        array = confusion_matrix(y_test, y_pred)
        df_cm = pd.DataFrame(array, index = self.category_list.unique(),
                  columns = self.category_list.unique())
        plt.figure(figsize = (10,7))
        sns.heatmap(df_cm, annot=True)
        print('Accuracy' + str(accuracy_score(y_test, y_pred)))
    
    def DTC(self):
        from sklearn.preprocessing import LabelEncoder
        from sklearn.tree import DecisionTreeClassifier
        from sklearn.model_selection import train_test_split
        from sklearn.metrics import confusion_matrix
        from sklearn.metrics import accuracy_score
        le = LabelEncoder()
        y = le.fit_transform(self.category_list.values)
        
        X_train, X_test, y_train, y_test = train_test_split(self.x, y, random_state=1)
        dt = DecisionTreeClassifier()
        dt.fit(X_train, y_train)
        y_pred = dt.predict(X_test)
        array = confusion_matrix(y_test, y_pred)
        df_cm = pd.DataFrame(array)#, index = self.category_list.unique(),
                  #columns = self.category_list.unique())
        plt.figure(figsize = (10,7))
        sns.heatmap(df_cm, annot=True)
        print('Accuracy' + str(accuracy_score(y_test, y_pred)))
    
    def QDA (self):
        from sklearn.discriminant_analysis import QuadraticDiscriminantAnalysis
        from sklearn.preprocessing import LabelEncoder
        from sklearn.tree import DecisionTreeClassifier
        from sklearn.model_selection import train_test_split
        from sklearn.metrics import confusion_matrix
        from sklearn.metrics import accuracy_score
        
        le = LabelEncoder()
        y = le.fit_transform(self.category_list.values)

        X_train, X_test, y_train, y_test = train_test_split(self.x, y, random_state=1)
        clf = QuadraticDiscriminantAnalysis()

        X_clf = clf.fit(self.x, y)
        
        y_pred = X_clf.predict(X_test)
        array = confusion_matrix(y_test, y_pred)
        df_cm = pd.DataFrame(array)#, index = self.category_list.unique(),
                  #columns = self.category_list.unique())
        plt.figure(figsize = (10,7))
        sns.heatmap(df_cm, annot=True)

    def K_means(self):

        from sklearn.cluster import KMeans


        df = pd.DataFrame(self.x, columns=self.category_list.unique())
        print(df.columns)
        kmeans = KMeans(n_clusters=11)
        print(type(self.category_list.unique(). tolist()))
        y = kmeans.fit_predict(df[df.columns])

        df['Cluster'] = y

        print(df.head())
        
    def CCA(self):
        from sklearn.preprocessing import LabelEncoder
        from sklearn.cross_decomposition import CCA
        
        from sklearn.metrics import confusion_matrix
        
        X = self.x.values
        
        n = X.shape[0]
        le = LabelEncoder()
        #Y = le.fit_transform(self.category_list_first, self.category_list_second)
        Y = pd.concat([pd.Series(le.fit_transform(self.category_list_first)), pd.Series(le.fit_transform(self.category_list_second))], axis=1)
        
        X_train = X[:n // 2]
        Y_train = Y[:n // 2]
        X_test = X[n // 2:]
        Y_test = Y[n // 2:]
        
        cca = CCA(n_components=2)
        cca.fit(X_train, Y_train)
        X_train_r, Y_train_r = cca.transform(X_train, Y_train)
        X_test_r, Y_test_r = cca.transform(X_test, Y_test)
        
        
        X_c, Y_c = cca.transform(X, Y)

        print(X_c.shape)
        print(Y_c.shape)
        
        cc_res = pd.DataFrame({"CCX_1":X_c[:, 0],
                       "CCY_1":Y_c[:, 0],
                       "CCX_2":X_c[:, 1],
                       "CCY_2":Y_c[:, 1],
                       "Typology":self.category_list_first.tolist(),
                       "Site":self.category_list_second.to_list()})
        import numpy as np
        np.corrcoef(X_c[:, 0], Y_c[:, 0])
        
        
        plt.figure(figsize=(10,8))
        sns.scatterplot(x="CCX_1",
                        y="CCY_1", 
                        hue="Typology", data=cc_res)
        plt.title('First Pair of Canonical Covariate, corr = %.2f' %
                 np.corrcoef(X_c[:, 0], Y_c[:, 0])[0, 1])
        plt.figure(figsize=(10,8))
        sns.scatterplot(x="CCX_2",
                        y="CCY_2", 
                        hue="Site", data=cc_res)
        plt.title('First Pair of Canonical Covariate, corr = %.2f' %
                 np.corrcoef(X_c[:, 1], Y_c[:, 1])[0, 1])
        
    def RDA(self):
        from sklearn.preprocessing import LabelEncoder
        from sklearn.cross_decomposition import RDA
        
        from sklearn.metrics import confusion_matrix
        
        X = self.x.values
        
        n = X.shape[0]
        le = LabelEncoder()
        #Y = le.fit_transform(self.category_list_first, self.category_list_second)
        Y = pd.concat([pd.Series(le.fit_transform(self.category_list_first)), pd.Series(le.fit_transform(self.category_list_second))], axis=1)
        
        X_train = X[:n // 2]
        Y_train = Y[:n // 2]
        X_test = X[n // 2:]
        Y_test = Y[n // 2:]
        
        cca = CCA(n_components=2)
        cca.fit(X_train, Y_train)
        X_train_r, Y_train_r = cca.transform(X_train, Y_train)
        X_test_r, Y_test_r = cca.transform(X_test, Y_test)
        
        
        X_c, Y_c = cca.transform(X, Y)

        print(X_c.shape)
        print(Y_c.shape)
        
        cc_res = pd.DataFrame({"CCX_1":X_c[:, 0],
                       "CCY_1":Y_c[:, 0],
                       "CCX_2":X_c[:, 1],
                       "CCY_2":Y_c[:, 1],
                       "Typology":self.category_list_first.tolist(),
                       "Site":self.category_list_second.to_list()})
        import numpy as np
        np.corrcoef(X_c[:, 0], Y_c[:, 0])
        
        
        plt.figure(figsize=(10,8))
        sns.scatterplot(x="CCX_1",
                        y="CCY_1", 
                        hue="Typology", data=cc_res)
        plt.title('First Pair of Canonical Covariate, corr = %.2f' %
                 np.corrcoef(X_c[:, 0], Y_c[:, 0])[0, 1])
        plt.figure(figsize=(10,8))
        sns.scatterplot(x="CCX_2",
                        y="CCY_2", 
                        hue="Site", data=cc_res)
        plt.title('First Pair of Canonical Covariate, corr = %.2f' %
                 np.corrcoef(X_c[:, 1], Y_c[:, 1])[0, 1])
        
        '''
        print(Y_mc)
        ca = CCA(n_components=3)
        ca.fit(X_mc, Y_mc)
        X_c, Y_c = ca.transform(X_mc, Y_mc)
        
        print(X_c.shape)
        print(Y_c.shape)
        
        cc_res = pd.DataFrame({"CCX_1":X_c[:, 0],
                       "CCY_1":Y_c[:, 0],
                       "CCX_2":X_c[:, 1],
                       "CCY_2":Y_c[:, 1],
                       "Typology":self.category_list.unique().tolist()})
        
        import numpy as np
        np.corrcoef(X_c[:, 0], Y_c[:, 0])
        
        sns.set_context("talk", font_scale=1.2)
        plt.figure(figsize=(10,8))
        sns.scatterplot(x="CCX_1",
                        y="CCY_1", 
                        hue="Species", data=cc_res)
        plt.title('First Pair of Canonical Covariate, corr = %.2f' %
                 np.corrcoef(X_c[:, 0], Y_c[:, 0])[0, 1])
        
        plt.figure(figsize=(10,8))
        sns.boxplot(x="Typology",
                        y="CCX_1", 
                       data=cc_res)
        sns.stripplot(x="Typology",
                        y="CCX_1", 
                         data=cc_res)
        
        plt.figure(figsize=(10,8))
        sns.boxplot(x="Typology",
                        y="CCY_1", 
                       data=cc_res)
        sns.stripplot(x="Typology",
                        y="CCY_1", 
                         data=cc_res)
        '''
        
    def R_RDA (self):
        import math, datetime
        import pandas as pd
        import numpy as np
        import os
        import tzlocal
        os.environ['R_HOME'] = 'C:\\Program Files\\R\\R-4.1.0'

        from rpy2.robjects import pandas2ri
        pandas2ri.activate()

        import subprocess

        import rpy2.robjects as robjects
        r = robjects.r
        r.source('../R/RDA.R')
        r['options'](warn=-1)
        
        
        from rpy2.robjects.conversion import localconverter

        rda = robjects.globalenv['rda']

        filter_list = np.array (['ID', 'tip_angle','ka_cal_BP','typology'])

        datfram = pd.read_csv('../../../ArchaeoCod/Outline_code/01__Gravetten/data_metrics_merged_filtered_typchr.csv')



        with localconverter(robjects.default_converter + pandas2ri.converter):
            r_from_pd_df = robjects.conversion.py2rpy(datfram)



        rda(r_from_pd_df,filter_list)
        pandas2ri.deactivate()
        
    def GMM (self):
        
        x = 1
```

```python
from sklearn.cluster import KMeans
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

cls = KMeans(init='k-means++', n_clusters=14, n_init=10)

df = dendro_pca_1.y

cls.fit(df.iloc[: , :-2])

df['predicted_label'] = cls.labels_.astype(int)
sns.scatterplot(x="PC1", 
                y="PC2", 
                hue="predicted_label", 
                style = 'typology',
                data=df, palette='Paired', s=20)
plt.legend(loc='lower right')
```
